```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load your cleaned dataset
df = pd.read_csv("C:\\Users\\trebo\\Desktop\\BDAActivity\\clean_movies.csv")
```

```python
# matplotlib Diagrams
# 1. Count movies per genre (Bar Graph)
genre_counts = df['genre'].value_counts()
plt.figure(figsize=(10, 5))
plt.bar(genre_counts.index, genre_counts.values, color='#87CEEB')
plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Movie Count per Genre")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Top 5 highest-grossing movies (Bar Chart)
top_grossing = df.nlargest(5, 'gross')
plt.figure(figsize=(10, 5))
plt.barh(top_grossing['name'], top_grossing['gross'], color='#87CEEB')
plt.xlabel("Gross Earnings")
plt.ylabel("Movie Name")
plt.title("Top 5 Highest-Grossing Movies")
plt.gca().invert_yaxis()  # Invert y-axis for better visualization
plt.show()
```

```python
# 3. Displays the distribution of movie scores (Histogram)
plt.figure(figsize=(8, 5))
plt.hist(df['score'], bins=20, color='#87CEEB', edgecolor='black')
plt.xlabel("Score")
plt.ylabel("Count of Movies")
plt.title("Distribution of Movie Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 4. Gross Earnings vs. Budget (Scatter Plot)
plt.figure(figsize=(8, 5))
plt.scatter(df['budget'], df['gross'], color='red', alpha=0.5)
plt.xlabel("Budget")
plt.ylabel("Gross Earnings")
plt.title("Gross Earnings vs. Budget")
plt.xscale('log')  
plt.yscale('log')
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
# 5. Count movies per year (Line Graph)
movies_per_year = df['year'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
plt.plot(movies_per_year.index, movies_per_year.values, marker='X', linestyle='-', color='#87CEEB')
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.title("Movies Released Per Year")
plt.grid(True)
plt.show()
```

```python
# Seaborn Diagram
# Set Seaborn style
sns.set_style("whitegrid")

# 1. Average Runtime vs. Score (Line Plot)
plt.figure(figsize=(10, 5))
runtime_score = df.groupby('score')['runtime'].mean().reset_index()
sns.lineplot(data=runtime_score, x='score', y='runtime', marker="X", color="#87CEEB")
plt.title("Average Runtime in Relation to Score")
plt.xlabel("Score")
plt.ylabel("Average Runtime (minutes)")
plt.show()
```

```python
# 2. Average Budget and Gross per Genre (Grouped Bar Plot)
plt.figure(figsize=(12, 6))
genre_budget_gross = df.groupby('genre')[['budget', 'gross']].mean().reset_index()
genre_budget_gross = genre_budget_gross.melt(id_vars="genre", var_name="Category", value_name="Amount")

sns.barplot(data=genre_budget_gross, x="genre", y="Amount", hue="Category", palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Average Budget and Gross Earnings per Genre")
plt.xlabel("Genre")
plt.ylabel("Amount (in millions)")
plt.show()
```

```python
# 3. Top 10 Artists with their respective average score (Bar Chart)
plt.figure(figsize=(10, 6))
top_stars = df['star'].value_counts().nlargest(10).index
avg_scores = df[df['star'].isin(top_stars)].groupby('star')['score'].mean().sort_values(ascending=False)
sns.barplot(x=avg_scores.values, y=avg_scores.index, palette="Blues_r")
plt.xlabel("Average Score")
plt.ylabel("Star")
plt.title("Top 10 Movie Stars with Their Average Score")
plt.show()
```

```python
# 4. Top 5 Directors by Average Movie Score (Box Plot)
top_5_directors = df.groupby('director')['score'].mean().nlargest(5).index
df_top_directors = df[df['director'].isin(top_5_directors)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_top_directors, x='director', y='score', palette="Set3")
plt.title("Top 5 Directors by Average Movie Score")
plt.xlabel("Director")
plt.ylabel("Score")
plt.xticks(rotation=25)
plt.show()
```

```python
# 5. Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['budget', 'gross', 'votes', 'score']].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Movie Attributes")
plt.show()
```
